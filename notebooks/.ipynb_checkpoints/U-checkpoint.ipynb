```python
import matplotlib.pyplot as plt
from nomic.atlas import map_data
import numpy as np
import pandas as pd
```

```python
u1 = np.random.uniform(size=100000)
u2 = np.random.uniform(size=100000)
u3 = np.random.uniform(size=100000)
```

```python
product = u1*u2
result = product**u3
```

```python
fig, ax = plt.subplots(1, 5, figsize=(25,3))
ax[0].hist(u1, bins=100)
ax[1].hist(u2, bins=100)
ax[2].hist(product, bins=100)
ax[3].hist(u3, bins=100)
ax[4].hist(result, bins=100)
plt.show()
```

```python
df = pd.DataFrame({
    "U1": u1,
    "U2": u2,
    "U3": u3,
    "U1*U2": product,
    "(U1*U2)^U3": result,
    "nickname": ["float" for _ in range(len(u1))]
})
map_data(data=df, indexed_field="nickname", identifier="unif", description="https://x.com/3blue1brown/status/1833534452187664468")
```
